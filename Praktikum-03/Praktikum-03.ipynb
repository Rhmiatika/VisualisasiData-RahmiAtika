```python
import matplotlib.pyplot as plt

# Data Sampel
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
product_A_sales = [10, 20, 15, 25, 30, 45, 40, 50, 60, 55, 65, 70]
product_B_sales = [5, 10, 8, 15, 18, 20, 22, 30, 25, 35, 40, 45]

# Line Chart
plt.plot(month, product_A_sales, label='Product A', color='Blue', linestyle='--', marker='o')
plt.plot(month, product_B_sales, label='Product B', color='red', linestyle='-', marker='x')
plt.title('Penjualan Produk Per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

JENIS GARIS UNTUK MENUNJUKKAN TREND

```python
import matplotlib.pyplot as plt

# Data Sampel
product_C_sales = [18, 22, 25, 28, 32, 38, 42, 45, 48, 52, 56, 60]
product_D_sales = [7, 9, 11, 13, 16, 18, 20, 23, 25, 28, 30, 35]

#   Plot Product A (dashed)
plt.plot(month, product_A_sales, label='Product A', color='Blue', linestyle='--')

# Plot Product B (solid)
plt.plot(month, product_B_sales, label='Product B', color='red', linestyle='-')

# Plot Product C (dash-dot)
plt.plot(month, product_C_sales, label='Product C', color='green', linestyle=':')

# Plot Product D (dotted)
plt.plot(month, product_D_sales, label='Product D', color='purple', linestyle='-.')

# Line Chart
plt.title('Penjualan Produk Per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# subplot, 2 baris 1 kolom, panjang 100 inch, tinggi 8 inch
fig, axs = plt.subplots(2, 1, figsize=(10,8))

# plot pertama untuk product C
axs[0].plot(month, product_C_sales, label='Product C', color='green', marker='d')
axs[0].set_title('Penjualan Produk C Per Bulan')
axs[0].set_xlabel('Bulan')
axs[0].set_ylabel('Jumlah Penjualan')
axs[0].legend()
axs[0].grid(True)

# plot pertama untuk product D
axs[1].plot(month, product_D_sales, label='Product D', color='purple', marker='s')
axs[1].set_title('Penjualan Produk D Per Bulan')
axs[1].set_xlabel('Bulan')
axs[1].set_ylabel('Jumlah Penjualan')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
